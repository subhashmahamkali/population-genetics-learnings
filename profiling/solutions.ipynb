## Problem 1.1: Genotype and Allele Frequencies

```python
# Define genotype counts
MM = 1787

MN = 3039
NN = 1303
total = MM + MN + NN
print(total)

#genotype frequencies
f_MM = MM / total
f_MN = MN / total
f_NN = NN / total

print(f_MM)
print(f_MN)
print(f_NN)

print(f_MM + f_MN + f_NN)

#allele frequencies
M = (2*MM + MN) / (2*total)
N = (2*NN + MN) / (2*total)

print(M)
print(N)
print(M + N)
```

```python
# Expected genotype frequencies under HWE
# M squre, 2 * M * N, N square

exp_f_MM = M**2
exp_f_MN = 2 * M * N
exp_f_NN = N**2 
print(exp_f_MM)
print(exp_f_MN)
print(exp_f_NN) 



```

```python
#How well do the observed genotype frequencies 
# match the expected genotype frequencies under HWE?
# Chi-square test
chi_square_MM = (f_MM - exp_f_MM)**2 / exp_f_MM
chi_square_MN = (f_MN - exp_f_MN)**2 / exp_f_MN
chi_square_NN = (f_NN - exp_f_NN)**2 / exp_f_NN
chi_square_total = chi_square_MM + chi_square_MN + chi_square_NN
print(chi_square_total)

M_totaal_from_exp = exp_f_MM * total
print(M_totaal_from_exp)
N_totaal_from_exp = exp_f_NN * total
print(N_totaal_from_exp)
MN_totaal_from_exp = exp_f_MN * total
print(MN_totaal_from_exp)

since the chi-square value is very high, we can conclude that 
the observed genotype frequencies do not match the 
expected genotype frequencies under HWE.
and can reject the null hypothesis that the population is in HWE.
```
